```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples

# Note: The result provided in your example 2 is incorrect

# you can free to change all these codes/structure
# here A and B are list of lists
def matrix_mul(A, B):
    matrixARowSize = len(A)
    if 0 == matrixARowSize:
        return
    
    matrixBRowSize = len(B)
    if 0 == matrixBRowSize:
        return
    
    matrixAColSize = len(A[0])
    if 0 == matrixAColSize:
        return
    
    matrixBColSize = len(B[0])
    if 0 == matrixBColSize:
        return
    
    if matrixAColSize != matrixBRowSize:
        print("Not Possible\n")
        return
    
    productList = list()
    for i in range(matrixARowSize):
        elementList = list()
        for j in range(matrixBColSize):
            product = 0
            for k in range(matrixAColSize):
                product += A[i][k] * B[k][j]
            elementList.append(product)
        productList.append(elementList)
    
    return productList
    
A = [[1, 2], [3, 4], [5, 6]]
B = [[1, 2, 3], [4, 5, 6]]
product = matrix_mul(A, B)
print(product)
```

<h3> Q5: Find the closest points</h3>

consider you have given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P
<br>cosine distance between two points (x,y) and (p,q) is defind as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples
# you can free to change all these codes/structure


# here S is list of tuples and P is a tuple ot len=2
def closest_points_to_p(S, P):
    referenceLst = list(P);
    
    closestPointStore = dict();
    finalList = []
    
    for i in range(len(S)):
        tupleLst = list(S[i])        
        numerator = (tupleLst[0] * referenceLst[0]) + (tupleLst[1] * referenceLst[1])
        denominator = (math.sqrt(tupleLst[0] ** 2 + tupleLst[1] ** 2)) * (math.sqrt(referenceLst[0] ** 2 + referenceLst[1] ** 2))
        result = numerator / denominator
        distance = math.acos(result)
        
        closestPointStore[distance] = i
    
    keysList = list(closestPointStore.keys())
    keysList.sort()
    
    for i in range(5):
        index = closestPointStore.get(keysList[i])
        
        if index == None:
            continue;
        
        finalList.append(S[index])
    
    return finalList

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
points = closest_points_to_p(S, P)
print(points) #print the returned values
```

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings


# you can free to change all these codes/structure

def curve_smoothing(string):
    newString = string.replace('_', '0')
    strList = newString.split(',')
       
    if 1 == len(strList) - strList.count('0'):
        for i in range(len(strList)):
            if '0' == strList[i]:
                continue
            else:
                result = int(strList[i]) / len(strList)
                break
        
        for i in range(len(strList)):
            strList[i] = result
        return strList
    
    left = 0
    right = 0

    for i in range(1, len(strList)):
        if i == len(strList) - 1:
            result = (int(strList[left]) + int(strList[i])) / (i - left + 1)
            for j in range(left, i + 1):
                strList[j] = result
            continue
            
        elif '0' == strList[i]:
            continue
        else:
            right = i
        result = (int(strList[left]) + int(strList[right])) / (right - left + 1)
        for j in range(left, i + 1):
            strList[j] = result
        left = right
    return strList

S = "_,_,_,24"
smoothed_values = curve_smoothing(S)
print(smoothed_values)

S = "40,_,_,_,60"
smoothed_values = curve_smoothing(S)
print(smoothed_values)

S = "80,_,_,_,_"
smoothed_values = curve_smoothing(S)
print(smoothed_values)

S = "_,_,30,_,_,_,50,_,_"
smoothed_values = curve_smoothing(S)
print(smoothed_values)
```

consider its like a martrix of n rows and two columns

a. the first column Y will contain interger values <br>
b. the second column $Y_{score}$ will be having float values <br>
Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.4243099
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
import math
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings

# you can free to change all these codes/structure
def compute_log_loss(A):
    product = 0
    for logLossList in A:
        product += (logLossList[0] * math.log(logLossList[1], 10)) + ((1 - logLossList[0]) * math.log(1 - logLossList[1], 10))
    
    loss = (product * -1) / len(A)
        
    return loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print(loss)
```
